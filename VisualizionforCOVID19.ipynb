# Creating data visualiztion for Cvoid-19 cases and deathes by country

```python
import numpy as np
import pandas as pd
import csv
import datetime
```

# data pre-processing

```python
# input data from time-series-19-covid-combined.csv
fpath = r'C:\P\VisualizationForCovid-19\data\time-series-19-covid-combined.csv'
csvFile = open(fpath)
reader = csv.reader(csvFile)

result = []
for item in reader:
    if reader.line_num == 1: # skip first line
        continue
    result.append(item)
    
csvFile.close()
#result
for item in result:
    for i in range(3,6):
        if item[i] != '':
            item[i] = int(item[i])
        else:
            item[i] = 0

# using DataFrame
df = pd.DataFrame(data = result,columns=['Time','Country/Region','Province/State','Confirmed','Recovered','Deaths'])
# string to time
df['Time']=pd.to_datetime(df['Time'])
# clean none data
df.dropna()
df
```

# Count how many days and countries/regions

```python
countries = len(df['Country/Region'].unique())
days = len(df['Time'].unique())
```

# Creating three DataFrame to store data for confirmed, recovered and death

```python
df_C = pd.DataFrame(data = [[0]*countries]*days, index=df['Time'].unique(), columns=df['Country/Region'].unique())
df_R = pd.DataFrame(data = [[0]*countries]*days, index=df['Time'].unique(), columns=df['Country/Region'].unique())
df_D = pd.DataFrame(data = [[0]*countries]*days, index=df['Time'].unique(), columns=df['Country/Region'].unique())
```

```python
grouped = df.groupby(by=['Country/Region','Time']).sum()
grouped
```

```python
for i in range(countries):
    data1 = list(grouped.iloc[ days*i:days*(i+1),0])
    data2 = list(grouped.iloc[ days*i:days*(i+1),1])
    data3 = list(grouped.iloc[ days*i:days*(i+1),2])
    df_C.iloc[0:days,i] = data1
    df_R.iloc[0:days,i] = data2
    df_D.iloc[0:days,i] = data3
    
df_C = df_C.T
df_R = df_R.T
df_D = df_D.T
```

# Output data

```python
outpathC = r'C:\P\VisualizationForCovid-19\Outputdata\Cases_byCountries.csv'
outpathR = r'C:\P\VisualizationForCovid-19\Outputdata\Recover_byCountries.csv'
outpathD = r'C:\P\VisualizationForCovid-19\Outputdata\Death_byCountries.csv'
df_C.to_csv(outpathC)
df_R.to_csv(outpathR)
df_D.to_csv(outpathD)

```
